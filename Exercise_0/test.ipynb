```python
from molsim import MonteCarlo
import numpy as np
import matplotlib.pyplot as plt
```

```python
!pip freeze | grep -i mol
```

```python
# Set up the Monte Carlo simulation
mc = MonteCarlo(
    numberOfParticles=100,
    temperature=1.0,
    boxSize=6.0,
    maxDisplacement=0.5,
    numberOfInitCycles=0,
    numberOfProdCycles=1000,
    sampleFrequency=100,
    seed=12,
)

# Run the Monte Carlo simulation
mc.run()
```

```python
# Plot the results
plt.plot(mc.energies, label="Potential energy")

plt.xlabel("Cycles")
plt.ylabel(r"Potential energy, U / $\varepsilon$")

plt.tight_layout()
```
